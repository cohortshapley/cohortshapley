# Evaluate unrealistic sampling of joint marginal distribution

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from cohortshapley import dataset
from cohortshapley import realistic
from cohortshapley import sampling
from cohortshapley import similarity
```

```python
def realistic_sampling(tratios, trealistic_rates1, trealistic_rates2, trealistic_rates3, trealistic_rates_marginal):
    plt.ylabel('Ratio of realistic samples')
    plt.xlabel('Distance threshold (ratio in value range)')

    y=trealistic_rates1.mean(axis=0)
    y_l=trealistic_rates1.min(axis=0)
    y_h=trealistic_rates1.max(axis=0)
    plt.plot(tratios, y,label='Hold out 10%')
    plt.fill_between(tratios, y_l, y_h, alpha=.1)
    
    y=trealistic_rates2.mean(axis=0)
    y_l=trealistic_rates2.min(axis=0)
    y_h=trealistic_rates2.max(axis=0)
    plt.plot(tratios, y,label='Hold out 20%')
    plt.fill_between(tratios, y_l, y_h, alpha=.1)
    
    y=trealistic_rates3.mean(axis=0)
    y_l=trealistic_rates3.min(axis=0)
    y_h=trealistic_rates3.max(axis=0)
    plt.plot(tratios, y,label='Hold out 30%')
    plt.fill_between(tratios, y_l, y_h, alpha=.1)
    
    y=trealistic_rates_marginal.mean(axis=0)
    y_l=trealistic_rates_marginal.min(axis=0)
    y_h=trealistic_rates_marginal.max(axis=0)
    plt.plot(tratios, y,label='Marginal dist.')
    plt.fill_between(tratios, y_l, y_h, alpha=.1)
```

## Titanic

```python
X,Y,categorical = dataset.titanic()
tratios, trealistic_rates1, trealistic_rates2, trealistic_rates3, trealistic_rates_marginal = realistic.evaluate_similarity_ratio(X,Y,categorical,n_trial=10)
```

```python
fig = plt.figure(figsize=(4,2),dpi=300)
plt.rcParams["font.size"] = 9
realistic_sampling(tratios, trealistic_rates1, trealistic_rates2, trealistic_rates3, trealistic_rates_marginal)
plt.legend()
plt.tight_layout()
plt.show()
```
